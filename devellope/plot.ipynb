```python
# Cell 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Cell 2: Load Status Data
status_file = r'c:\daisyrobot\robomas\csv\status_data_20250702_123954.csv'
status_data = pd.read_csv(status_file)
status_data['timestamp'] = pd.to_datetime(status_data['timestamp'])
status_data['data'] = status_data['data'].apply(lambda x: eval(x))

# Cell 3: Plot Status Data
plt.figure(figsize=(10, 6))
for i in range(len(status_data['data'][0])):
    plt.plot(status_data['timestamp'], [d[i] for d in status_data['data']], label=f'Status {i}')
plt.xlabel('Timestamp')
plt.ylabel('Status Values')
plt.title('Status Data')
plt.legend()
plt.grid()
plt.show()




```

```python
# Cell 4: Load Position Data
position_file = r'c:\daisyrobot\robomas\csv\position_data_20250702_123954.csv'
position_data = pd.read_csv(position_file)
position_data['timestamp'] = pd.to_datetime(position_data['timestamp'])
position_data['data'] = position_data['data'].apply(lambda x: eval(x))

# Cell 5: Plot Position Data
plt.figure(figsize=(10, 6))
labels = ['X', 'Y', 'Z']
for i in range(len(position_data['data'][0])):
    plt.plot(position_data['timestamp'], [d[i] for d in position_data['data']], label=f'Position {labels[i]}')
plt.xlabel('Timestamp')
plt.ylabel('Position Values')
plt.title('Position Data')
plt.legend()
plt.grid()
plt.show()
```

```python
# Cell 6: Load IMU Data
imu_file = r'c:\daisyrobot\robomas\csv\imu_data_20250702_123954.csv'
imu_data = pd.read_csv(imu_file)
imu_data['timestamp'] = pd.to_datetime(imu_data['timestamp'])
imu_data['data'] = imu_data['data'].apply(lambda x: eval(x))

# Cell 7: Plot IMU Data (accelerometer and gyroscope)
plt.figure(figsize=(10, 6))
labels = ['Accel X', 'Accel Y', 'Accel Z', 'Gyro X', 'Gyro Y', 'Gyro Z']
for i in range(len(imu_data['data'][0])):
    plt.plot(imu_data['timestamp'], [d[i] for d in imu_data['data']], label=f'IMU {labels[i]}')
plt.xlabel('Timestamp')
plt.ylabel('IMU Values')
plt.title('IMU Data')
plt.legend()
plt.grid()
plt.show()

```
