```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ‡∏≠‡πà‡∏≤‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å‡πÑ‡∏ü‡∏•‡πå CSV
df = pd.read_csv('movement_log.csv')

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡πÄ‡∏™‡πâ‡∏ô‡∏ó‡∏≤‡∏á XY
plt.figure(figsize=(12, 8))

# ‡∏û‡∏•‡∏≠‡∏ï‡πÄ‡∏™‡πâ‡∏ô‡∏ó‡∏≤‡∏á
plt.plot(df['x'], df['y'], 'b-', linewidth=2, alpha=0.7, label='‡πÄ‡∏™‡πâ‡∏ô‡∏ó‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏Ñ‡∏•‡∏∑‡πà‡∏≠‡∏ô‡∏ó‡∏µ‡πà')

# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏à‡∏∏‡∏î‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡πÅ‡∏•‡∏∞‡∏à‡∏∏‡∏î‡∏™‡∏¥‡πâ‡∏ô‡∏™‡∏∏‡∏î
plt.plot(df['x'].iloc[0], df['y'].iloc[0], 'go', markersize=10, label=f'‡∏à‡∏∏‡∏î‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô ({df["x"].iloc[0]:.3f}, {df["y"].iloc[0]:.3f})')
plt.plot(df['x'].iloc[-1], df['y'].iloc[-1], 'ro', markersize=10, label=f'‡∏à‡∏∏‡∏î‡∏™‡∏¥‡πâ‡∏ô‡∏™‡∏∏‡∏î ({df["x"].iloc[-1]:.3f}, {df["y"].iloc[-1]:.3f})')

# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏•‡∏π‡∏Å‡∏®‡∏£‡πÅ‡∏™‡∏î‡∏á‡∏ó‡∏¥‡∏®‡∏ó‡∏≤‡∏á (‡∏ó‡∏∏‡∏Å 50 ‡∏à‡∏∏‡∏î)
for i in range(0, len(df), 50):
    if i+1 < len(df):
        dx = df['x'].iloc[i+1] - df['x'].iloc[i]
        dy = df['y'].iloc[i+1] - df['y'].iloc[i]
        plt.arrow(df['x'].iloc[i], df['y'].iloc[i], dx*10, dy*10, 
                 head_width=0.01, head_length=0.01, fc='red', ec='red', alpha=0.6)

# ‡∏ï‡∏Å‡πÅ‡∏ï‡πà‡∏á‡∏Å‡∏£‡∏≤‡∏ü
plt.xlabel('X Position (m)', fontsize=12)
plt.ylabel('Y Position (m)', fontsize=12)
plt.title('‡πÄ‡∏™‡πâ‡∏ô‡∏ó‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏Ñ‡∏•‡∏∑‡πà‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡∏´‡∏∏‡πà‡∏ô‡∏¢‡∏ô‡∏ï‡πå (Robot Movement Path)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.axis('equal')

# ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥
total_distance = np.sum(np.sqrt(np.diff(df['x'])**2 + np.diff(df['y'])**2))
print(f"üìä ‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÄ‡∏Ñ‡∏•‡∏∑‡πà‡∏≠‡∏ô‡∏ó‡∏µ‡πà:")
print(f"   - ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏à‡∏∏‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•: {len(df):,} ‡∏à‡∏∏‡∏î")
print(f"   - ‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á‡∏£‡∏ß‡∏°: {total_distance:.3f} ‡πÄ‡∏°‡∏ï‡∏£")
print(f"   - ‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏ß‡∏•‡∏≤: {df['time'].iloc[-1] - df['time'].iloc[0]:.1f} ‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ")
print(f"   - ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô: ({df['x'].iloc[0]:.3f}, {df['y'].iloc[0]:.3f})")
print(f"   - ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏™‡∏¥‡πâ‡∏ô‡∏™‡∏∏‡∏î: ({df['x'].iloc[-1]:.3f}, {df['y'].iloc[-1]:.3f})")

plt.tight_layout()
plt.show()

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏ï‡∏¥‡∏° - ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß‡∏ï‡∏≤‡∏°‡πÄ‡∏ß‡∏•‡∏≤
plt.figure(figsize=(12, 6))
plt.plot(df['time'] - df['time'].iloc[0], df['speed'], 'b-', linewidth=1, alpha=0.7)
plt.xlabel('‡πÄ‡∏ß‡∏•‡∏≤ (‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ)')
plt.ylabel('‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß (m/s)')
plt.title('‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß‡∏Å‡∏≤‡∏£‡πÄ‡∏Ñ‡∏•‡∏∑‡πà‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡∏ï‡∏≤‡∏°‡πÄ‡∏ß‡∏•‡∏≤')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ‡∏Å‡∏£‡∏≤‡∏ü PID values
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 10))

# Proportional
ax1.plot(df['time'] - df['time'].iloc[0], df['p'], 'r-', linewidth=1, alpha=0.7)
ax1.set_ylabel('P value')
ax1.set_title('PID Controller Values')
ax1.grid(True, alpha=0.3)

# Integral
ax2.plot(df['time'] - df['time'].iloc[0], df['i'], 'g-', linewidth=1, alpha=0.7)
ax2.set_ylabel('I value')
ax2.grid(True, alpha=0.3)

# Derivative
ax3.plot(df['time'] - df['time'].iloc[0], df['d'], 'b-', linewidth=1, alpha=0.7)
ax3.set_xlabel('‡πÄ‡∏ß‡∏•‡∏≤ (‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ)')
ax3.set_ylabel('D value')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
